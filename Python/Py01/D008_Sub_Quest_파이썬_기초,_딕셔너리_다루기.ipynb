## Q1-1. 회문 판독기

문자열을 입력으로 받으면 해당 문자열이 회문인지 아닌지 판독하여 회문인 경우 True, 아닌 경우 False를 반환하는 함수를 만듭니다.

이때, 반환되는 True or False는 문자열이 아닌 'bool' 타입입니다.

---
입력 예시:
```python
target1 = "아많다많다많다많아"
target2 = "아정말많다많다많아"

print(is_palindrome(target1))
print(is_palindrome(target2))
```

출력 예시:
```
True
False
```

````python
def is_palindrome(text):
  text_list = [x for x in text] # pop함수를 사용하기 위해 문자열은 리스트로 변경
  text_ud = ""                  # 문자열을 거꾸로 하여 저장할 변수를 선언

  i = 1
  while i <= len(text):
    text_ud += text_list.pop()
    i += 1

  return text_ud == text
````

````python
# Test
target1 = "여보게저기저게보여"
target2 = "보이긴뭐가보여저기"

print(is_palindrome(target1))
print(is_palindrome(target2))
````

## Q1-2. 회문 분류기

Q1-1에서 만든 판독기 함수를 이용하여 회문 분류기 함수를 만들어 봅니다.

회문 분류기 함수는 아래의 작업을 수행합니다.

1. 여러 단어가 포함된 리스트를 입력으로 받으면 한 단어씩 판독기를 통해 회문 여부를 확인 후 하나의 딕셔너리에 회문과 회문이 아닌 단어를 분류하여 저장합니다.
    - 이때, 회문의 경우 딕셔너리에서 key값은 "회문O", value값은 회문인 단어들로 이루어진 리스트입니다.
    - 회문이 아닌 경우의 key값은 "회문X", value값은 회문이 아닌 단어들로 이루어진 리스트입니다.

2. 모든 단어의 분류가 완료되면 저장한 딕셔너리와, 회문의 총 개수, 전체에서 회문의 비율을 계산하여 반환합니다.
    - 이때, 회문의 총 개수는 int, 회문의 비율은 float입니다.
    - 회문의 비율이 `30%`라면 반환되는 값은 `0.3`입니다.

---
예시 입력:
```python
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier(targets))
```

예시 출력:
```
({'회문O': ['기러기', '토마토', '스위스', '인도인', '별똥별', '우영우', '우향우'], '회문X': ['까마귀', '스윗소스', '한국인']}, 7, 0.7)
```

````python
# 기존 is_palindrome 함수 사용 하여 실행시 '기','러','기',... 로 글자가 분리되어 저장되는 문제를 해결하고자 딕셔너리로 반환되는 새로운 함수를 만듬

def is_palindrome_dict(text):
  text_list = [x for x in text]
  text_ud = ""

  i = 1
  while i <= len(text):
    text_ud += text_list.pop()
    i += 1

  dict_palindrom = {} # 리턴할 딕셔너리 변수 선언
  dict_palindrom.setdefault('text',text) # 판별된 bool 값 저장
  dict_palindrom.setdefault('upside_down',text_ud) # 회문
  dict_palindrom.setdefault('bool',text_ud == text)  # 판별된 bool 값 저장
  return dict_palindrom
````

````python
# is_palindrome_dict 함수 테스트
print(is_palindrome_dict('까마귀'))
````

````python
# is_palindrome_dict 함수를 활용한 회문 분류기
def palindrome_classifier_dict(targets):
    palindrome_y = [] # 회문으로 판독된 단어를 저장하는 리스트 변수 선언
    palindrome_n = [] # 회문이 아닌 것으로 판별된 단어를 저장하는 리스트 변수 선언
    palindrome_category = {} # 회문인 문장과, 회문이 아닌 문장을 Key 값과 함께 딕셔너리로 저장하기 위한 변수 선언

    for word in targets:
      result_dict = is_palindrome_dict(word) # is_palindrome_dict 함수를 호출
      if result_dict['bool']:
        palindrome_y.append(word)
      else:
        palindrome_n.append(word)

    palindrome_category.setdefault ('회문O',palindrome_y)
    palindrome_category.setdefault ('회문X',palindrome_n)

    palindrome_len = len(palindrome_y)
    palindrome_rate = int(palindrome_len / len(targets) * 100)
    return palindrome_category, palindrome_len, palindrome_rate
````

````python
# Test
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier_dict(targets))
````

## Q2-1. 성적 데이터화

하나의 변수 안에 다음의 시험 성적 데이터를 적절한 자료형의 객체로 저장하고자 합니다.  
이후 해당 변수는 성적 분석 함수에서 참조할 예정입니다.

예를 들어 변수명을 ‘scores’라고 했다면 ‘scores’안에 아래 성적 내용이 모두 담겨 있어야 합니다.

||국어|영어|수학|
|-|-|-|-|
|유원영|94|91|89|
|설지민|88|96|90|
|신윤아|86|88|77|
|장윤아|98|76|92|

## Q2-2. 성적 분석기

이제 Q2-1에서 만든 변수를 참조하여 학생의 이름을 인자로 받아 아래의 예시와 같이 성적을 분석해주는 함수를 만들어 봅니다.

---

예시 입력:
```python
score_analysis("설지민")
```
예시 출력:
```
[설지민 성적 분석]
국어: 88점, 학급 평균(91.50점) 대비 -3.50
영어: 96점, 학급 평균(87.75점) 대비 +8.25
수학: 90점, 학급 평균(87.00점) 대비 +3.00

전과목 평균: 91.33점, 학급 평균(88.75점) 대비 +2.58
```


````python
def score_analysis(name):
  report_scores = [['국어','영어','수학'],[],[]]
  i = 1
  while i < len(scores):
    report_scores[1].append(sum(scores[i]))
    report_scores[2].append(sum(scores[i])/len(scores[i]))
    i += 1
  j = scores[0].index(name)
  print(f'[{name} 성적 분석]')
  i = 1
  for subject in report_scores[0]:
    print(f'{subject}: {scores[i][j]}점, 학급 평균({round(report_scores[2][0],2)}점) 대비 {scores[i][j]-report_scores[2][0]}')
    i += 1

````

````python
# TEST
score_analysis("설지민")
````

````python
score_analysis("신윤아")
````
