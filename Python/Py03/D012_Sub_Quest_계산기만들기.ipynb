#  Quest C
### 사용자로부터 두 개의 정수와 연산자(+,-,*,/,**)를 입력받아 계산 결과를 출력하는 프로그램 만들어 보자

#### 조건
- 사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다
- 정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 다시 입력받기를 시도합니다.
- 나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.
- 사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.
- math 모듈을 사용하여 제곱 연산(**)을 지원합니다.
- 계산기를 통해 계속 계산을 할 것인지 입력받습니다.

## Pair Programming 결과

```python
# import math  # ** 대신 math 표준모듈을 사용하고자 할 때
```

```python
def calculator():
  # 계산계속 상태를 지정. q이면 계산기 작동을 종료
  # 일단 1회 시행해야하므로 'q'가 아닌 임의의 값을 넣음
  is_continue = 'y'

  while is_continue != 'q':
    user_input = input("수식을 입력하세요: ").strip()

    # 입력된 수식이 현재 계산기로 계산 가능한 연산자인지 확인
    target_op = None
    for operator in ['**', '+', '-', '/', '*']:
        if operator in user_input:
           target_op = operator
           index = user_input.find(operator)
           break
    if target_op == None:
      print("지원하지 않는 연산자입니다.")
      continue

    # 지원하는 연산자이면 계산을 시도
    try:
      # 먼저 연산자를 기준으로 수를 잘라서 변수에 저장
      num1 = int(user_input[:index])
      num2 = int(user_input[index + len(target_op):])
      # 연산자의 종류에 따라 계산 시도
      if target_op == '**':
        result = num1**num2
      elif target_op == '+':
        result = num1 + num2
      elif target_op == '-':
        result = num1 - num2
      elif target_op == '/':
        result = num1 / num2
      elif target_op == '*':
        result = num1 * num2
      else:
        # 계산이 불가능하다면 연산할 수 없는 글자라 판단 ValueError를 발생
        raise ValueError

    # 에러처리
    except ValueError:
      print("잘못된 입력입니다. 정수를 입력해주세요.")
    except ZeroDivisionError:
      print("0으로 나눌 수 없습니다.")

    # 계산시 에러가 없다면 계산 결과 반환
    else:
      print(f"계산결과: {result}")

    # 에러가 있든 없든, 마지막에는 계산을 더 할껀지 물어봄
    finally:
      is_continue = input("계산을 종료 하시겠습니까? q를 입력하세요.").strip()
      if is_continue == 'q':
        print("계산을 종료합니다.")
```

```python
calculator()
```
