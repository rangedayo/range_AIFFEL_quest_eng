# 1. 사각형 넓이 구하기
직사각형(rectangle), 평행사변형(parallelogram), 사다리꼴(trapezoid)의 넓이를 모두 구할 수 있는 클래스를 만들기
직사각형은 가로과 세로를 입력받아야 하고, 평행사변형은 밑변과 높이를, 사다리꼴은 밑변과 윗변 그리고 높이를 입력받아야 함
- 직사각형: 가로 x 세로
- 평행사변형: 밑변 x 높이
- (등변) 사다리꼴: (밑변 + 윗변) x 높이 x 1/2







## 1-1. 사각형 클래스 만들기  
세 가지 사각형의 넓이를 구할 수 있는 사각형(Square) 클래스를 만들기

### ✨최종 사각형 클래스 코드✨

#### 생성된 객체의 형식에 맞은 넓이 구하기 메소드만 실행되겠금 변경
- 생성된 사각형 형태에 따라 메소드 호출 제안하기
- '직사각형 < 평행사변형 < 사다리꼴' 이므로 직사각형은 모든 메소드 호출이 가능하지만 평행사변형은 사다리꼴까지만, 사다리꼴은 사다리꼴 메소드만 호출 가능하게!!

```python
class Square:
    def __init__(self):  # 생성자__init__의 매개변수의 첫번째는 self여야 합니다.
        self.attempt_limit = 3
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))  # 입력값을 int로 묶어야 합니다.
        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
            self.type = '직사각형(rectangle)'

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
            self.type = '평행사변형(parallelogram)'

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
            self.type = '사다리꼴(trapezoid)'

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')


    def rect(self):
        attempt = 0
        area = None
        if self.square == 1:
            while attempt < self.attempt_limit:
                try:
                    width, height = map(int,input('직사각형의 가로, 세로의 길이를 입력하세요. 예시: 20, 18 ' ).split(','))
                    area = width * height
                except:
                    print("입력값이 잘못되었습니다.  입력형식은 '가로, 세로'로 자연수만 입력가능합니다.. 다시 입력해 주세요")
                else:
                    result = '직사각형의 넓이는 : ' + str(area)
                    attempt = self.attempt_limit
                finally:
                    attempt += 1
        else:
            print("이 오브젝트는 직사각형이 아닙니다. 이 오브젝트의 유형을 알고 싶으시면 type 변수를 확인해주세요.")

        return result

    def par(self):
        attempt = 0
        area = None
        if self.square != 3:
            if self.square == 1:
              print("이 오브젝트는 직사각형입니다. 직사각형은 네각이 90도인 평행사변형이므로, 평행사변형의 넓이를 구하는 방식으로 직사각형의 넓이를 구할 수 있습니다.")
            while attempt < self.attempt_limit:
                try:
                    width, height = map(int,input('평행사변형의 밑변, 높이를 입력하세요. 예시: 20, 18 ').split(','))
                    area = width * height
                except:
                    print("입력값이 잘못되었습니다.  입력형식은 '밑변, 세로'로 자연수만 입력가능합니다.. 다시 입력해 주세요")
                    attempt += 1
                else:
                    result = '평행사변형의 넓이는 : ' + str(area)
                    attempt = self.attempt_limit
                finally:
                    attempt += 1
        else:
            print("이 오브젝트는 평행사변형이 아닙니다. 이 오브젝트의 유형을 알고 싶으시면 type 변수를 확인해주세요.")

        return result

    def trape(self):
        attempt = 0
        area = None
        if self.square == 1:
          print("이 오브젝트는 직사각형입니다. 직사각형도 사다리꼴에 포함되므로, 사다리꼴의 넓이를 구하는 방식으로 직사각형의 넓이를 구할 수 있습니다.")
        if self.square == 2:
          print("이 오브젝트는 평행사변형입니다. 평행사변형은 밑변과 윗변의 길이가 같은 사다리꼴이므로, 사다리꼴의 넓이를 구하는 방식으로 평행사변형의 넓이를 구할 수 있습니다.")
        while attempt < self.attempt_limit:
          try:
            bt_width, top_width, height = map(int,input('사다리꼴의 밑변, 윗변, 높이를 입력하세요. 예시: 14, 12, 21 ').split(','))
            area = ( bt_width + top_width ) * height * 0.5
          except:
            print("입력값이 잘못되었습니다.  입력형식은 '밑변, 윗변, 높이'로 자연수만 입력가능합니다.. 다시 입력해 주세요")
          else:
            result = '사다리꼴의 넓이는 : ' + str(area)
            attempt = 3
          finally:
            attempt += 1
        return result
```

## 2-1. 메뉴 생성 (1) 클래스 활용하기
- 클래스를 이용하여 메뉴 생성기 만들기
- 그러나 메뉴 여러개를 일일히 입력하려고 하면 너무 번거롭겠다 싶어서 PASS

```python
import sys

# 커피 메뉴와 가격을 생성하기 위한 틀, 메뉴 클래스 만들기
class MenuInfo:
  # 부정입력 방지를 위한 재시도 횟수 제한
  attempt_limit = 5
  # 카테고리 넘버, 카테고리 이름
  category_info = ['커피','논커피']
  # 옵션명, 옵션 금액
  option_info = [['개인텀블러사용',-1000],['아이스',0],['디카페인',300],['사이즈업',800],['휘핑크림추가',800],['샷추가',800]]
  # 메뉴를 저장하기 위한 공 리스트
  menu_list = []

  def __init__(self):

    # 카테고리 안내 문구 생성하기
    i = 0
    self.category_guide = "메뉴를 등록할 카테고리를 입력하세요 \n"
    while i < len(self.category_info):
      self.category_guide += str(i+1) + '. ' + self.category_info[i] + '\n'
      i += 1

    # 옵션 안내 문구 생성하기
    i = 0
    self.option_guide = "해당 메뉴에 추가 가능한 옵션을 선택해주세요. \n"
    while i < len(self.option_info):

      self.option_guide += str(i+1) + '. ' + self.option_info[i][0] + '　'*(15 - len(self.option_info[i][0])) + ' '*(10 - len(str(self.option_info[i][1]))) + str(self.option_info[i][1]) +'원\n'
      i += 1

    # 카테고리 입력 받기
    attempt = 0
    while attempt < self.attempt_limit:
      print(self.category_guide)
      try:
        self.m_category = int(input("숫자만 입력하세요 >>> "))
        if 1 <= self.m_category <= len(self.category_info):
          break
        else:
          print('입력이 올바르지 않습니다. 다시 입력해주세요')
      except ValueError:
        print('잘못된 입력입니다. 숫자를 입력해주세요.')
      attempt += 1

    if attempt >= self.attempt_limit:
      print("입력오류가 허용 횟수를 초과하였습니다. 처음부터 다시 시도해주세요.")
      sys.exit()

    # 메뉴 입력받기
    attempt = 0
    self.m_name = None
    self.m_price = 0
    attempt = 0 # reset attempt counter
    while attempt < self.attempt_limit:
      try:
        if self.m_name == None :
          self.m_name = str(input("메뉴판에 표시할 메뉴명을 입력해주세요 >>> "))
        if self.m_price ==  0 :
          self.m_price = int(input("해당 메뉴의 가격을 입력해주세요 >>> "))
        break # Exit loop if inputs are valid
      except ValueError:
        print("잘못된 입력입니다. 가격은 숫자로 입력해주세요.")
      attempt += 1
    if attempt >= self.attempt_limit:
      print("입력오류가 허용 횟수를 초과하였습니다. 처음부터 다시 시도해주세요.")
      sys.exit()

    # 해당 메뉴에 선택 가능한 옵션 입력하기
    attempt = 0
    self.m_option = []
    print(self.option_guide)
    while attempt < self.attempt_limit:
      try:
        self.m_option = map(int,input('추가하고 싶은 옵션을 쉼표로 구분해서 입력해주세요. 예) 1,3,4').split(','))
      except:
        print("잘못된 입력입니다. 쉼표로 구분해서 입력해주세요.")
        attempt += 1
        if attempt >= self.attempt_limit:
          print("입력오류가 허용 횟수를 초과하였습니다. 처음부터 다시 시도해주세요.")
          sys.exit()
      else:
        break
```

## 2-2. 메뉴판 만들기 (딕셔너리로)

#### **[코드 작성]**

```python
def printMn():
  print('─' * 20)
  print('　' * 8 + ' 메　뉴')
  print('─' * 20)

  mn_code_list = []

  for ct_id in category_info:
    ct_name = category_info[ct_id][0]
    ct_option = category_info[ct_id][1]
    print(ct_name)
    print('- '*20)

    j = 0
    while j < len(menu_list[ct_id]):
      mn_name, mn_price, mn_option = menu_list[ct_id][j]
      mn_temp=''
      # 음료 온도 옵션 저장
      for k in [0,1]:
        if k in mn_option:
          if len(mn_temp) > 0:
            mn_temp += '/'
          mn_temp += option_info[k][0]
      if len(mn_temp) > 0:
        mn_temp = '(' + mn_temp + ')'
      # 메뉴 출력: 음료명(Hot/Ice), 가격
      print(str(ct_id) + str(j+1) + '. ' + mn_name + mn_temp + ' ' * max(0, 25 - len(mn_name) - len(mn_temp)) + ' ' * max(0, 10 - len(str(mn_price))) + str(mn_price) )
      mn_code_list.append(int(str(ct_id) + str(j+1)))
      # 메뉴 단품당 추가옵션이 있다면 바로 하단에 안내 출력
      mn_guide = ''
      for k in mn_option:
        if k not in [0,1]: # Ensure k is not 'Hot' or 'Ice' if handled by mn_temp
          if len(mn_guide) > 0:
            mn_guide += ' / '
          mn_guide += option_info[k][0]
      if len(mn_guide) > 0:
        mn_guide = '    + ' + mn_guide
        print(mn_guide)
      j += 1

    # 메뉴 카테고리 공동 옵션이 있다면 카테고리 종료 후 바로 출력
    ct_guide = ''
    for m in ct_option:
      ct_guide += option_info[m][0]
      ct_guide += ' '
    if len(ct_guide) > 0:
      print('\n + ' + ct_guide)

    print('─' * 20)


  # 옵션 가격 안내
  option_guide =''
  for n in option_info:
    if n != 0:
      option_guide += str(n)+') '
      option_guide += option_info[n][0] + ' '*(30 - len(option_info[n][0]))
      option_guide += ' +' + ' '*(5 - len(str(option_info[n][1])))
      option_guide += str(option_info[n][1])
      option_guide += '\n'
  if len(option_guide) > 0:
    print('Option Price Guide && Order Number')
    print('- '*20)
    print(option_guide)
  print('─' * 20)
  return mn_code_list
```

```python
from random import random
from datetime import datetime
import unicodedata

def lenstr(s):
    """한글/영문 혼합 문자열의 실제 출력 너비를 계산"""
    tmp = str(s)
    return sum(2 if unicodedata.east_asian_width(c) in 'WF' else 1 for c in tmp)

class Kiosk:

  ## 생성자
  def __init__(self, category, option, menu):
    self.orderhistory = []
    self.idx =0
    self.category = category
    self.option = option
    self.menu = menu

    print("[카테고리] \n" + str(self.category))
    print("[옵션] \n" + str(self.option))
    print("[메뉴] \n" + str(self.menu))

    print("\n" + "메뉴판 생성이 완료되었습니다")

  ## 메뉴판 출력
  def menu_print(self):
    print('─' * 20)
    print('　' * 8 + ' 메　뉴')
    print('─' * 20)

    self.menu_num = []

    for ct_id in self.category:
      ct_name = self.category[ct_id][0]
      ct_option = self.category[ct_id][1]
      print(ct_name)
      print('- '*20)

      j = 0
      while j < len(self.menu[ct_id]):
        mn_name, mn_price, mn_option = self.menu[ct_id][j]
        mn_temp=''
        # 음료 온도 옵션 저장
        for k in [0,1]:
          if k in mn_option:
            if len(mn_temp) > 0:
              mn_temp += '/'
            mn_temp += option_info[k][0]
        if len(mn_temp) > 0:
          mn_temp = '(' + mn_temp + ')'
        # 메뉴 출력: 음료명(Hot/Ice), 가격
        print(str(ct_id) + str(j+1) + '. ' + mn_name + mn_temp + ' ' * max(0, 25 - lenstr(mn_name) - len(mn_temp)) + ' ' * max(0, 10 - len(str(mn_price))) + str(mn_price) )
        self.menu_num.append(int(str(ct_id) + str(j+1)))
        # 메뉴 단품당 추가옵션이 있다면 바로 하단에 안내 출력
        mn_guide = ''
        for k in mn_option:
          if k not in [0,1]: # Ensure k is not 'Hot' or 'Ice' if handled by mn_temp
            if lenstr(mn_guide) > 0:
              mn_guide += ' / '
            mn_guide += option_info[k][0]
        if lenstr(mn_guide) > 0:
          mn_guide = '    + ' + mn_guide
          print(mn_guide)
        j += 1

      # 메뉴 카테고리 공동 옵션이 있다면 카테고리 종료 후 바로 출력
      ct_guide = ''
      for m in ct_option:
        ct_guide += option_info[m][0]
        ct_guide += ' '
      if lenstr(ct_guide) > 0:
        print('\n + ' + ct_guide)

      print('─' * 20)

    # 옵션 가격 안내
    option_guide =''
    for n in option_info:
      if n != 0:
        option_guide += str(n)+') '
        option_guide += option_info[n][0] + ' '*(30 - lenstr(option_info[n][0]))
        option_guide += ' +' + ' '*(5 - lenstr(str(option_info[n][1])))
        option_guide += str(option_info[n][1])
        option_guide += '\n'
    if lenstr(option_guide) > 0:
      print('Option Price Guide && Order Number')
      print('- '*20)
      print(option_guide)
    print('─' * 20)
    return self.menu_num


  ## 주문하기
  def menu_select(self):
    self.orderlist = []  # [[메뉴번호,[옵션1,옵션2],수량], ... ]
    self.orderprice = []
    self.ordertotal = 0
    # 주문 메뉴와 옵션 선택하기
    i = 0
    is_end = 'n'
    while is_end != 'y':
      # 변수 초기화
      tmp_option = []   # 선택 가능한 옵션을 임시로 저장할 리스트
      tmp_select = []   # 선택한 옵션 초기화
      tmp_guide = ''# 옵션 안내문 초기화
      price_basic = 0 # 음료 기본가격
      price_option = 0 # 음료 옵션 가격

      print('위의 메뉴판을 보고 주문할 메뉴 번호를 입력해주세요.')
      print("(※ 예시: 아이스 아메리카노: '11' 입력)")

      try:
        mn_num = int(input("주문할 음료 번호를 입력하세요 >>> "))
        # 입력된 메뉴가 주문 가능한 메뉴가 맞는 지 검사
        if mn_num in self.menu_num:
          # 주문 가능한 메뉴가 맞다면 가격과 선택가능한 옵션 불러오기
          # 메뉴번호를 분리시켜 카테코리 키워드와 메뉴 키워드 찾기
          ct_key = int(str(mn_num)[0])
          mn_key = int(str(mn_num)[1:]) - 1 # 0-based 인덱스로 변경
          # 음료가격을 확인 기본가격에 넣기
          price_basic = int(self.menu[ct_key][mn_key][1])

          # 카테고리 공통 옵션 및 단품 옵션 값을 모아 선택가능한 옵션 리스트 만들기
          tmp_option.extend(self.category[ct_key][1])
          # 0-based 인덱스를 사용하도록 수정
          tmp_option.extend(self.menu[ct_key][mn_key][2])
          set_tmp_option = set(tmp_option)
          tmp_option = sorted(list(set_tmp_option)) # 정렬된 리스트로 변환

          print('+'*5)

          for j in tmp_option:
            # 옵션 0 ('Hot')은 일반적으로 표시하지 않으므로 건너뜁니다.
            if j != 0:
              tmp_guide += str(j) + ') '
              tmp_guide += self.option[j][0] + '(+' + str(self.option[j][1]) + ')' + '\n'

          # 선택한 옵션 입력받기
          if lenstr(tmp_option) > 0:
            print('해당 메뉴에 대한 선택 가능한 옵션입니다.')
            print(tmp_guide)
            print('선택할 옵션을 입력해주세요. 여러가지 옵션을 선택하고자 하는 경우 쉼표로 구분해주세요')
            print("(※ 예시: 아이스, 사이즈업, 샷추가 선택 : '1,3,4' 입력)")
            input_option_str = input('옵션 번호를 입력해주세요. >>> ' )
            if input_option_str: # 입력이 비어있지 않은 경우에만 처리
                input_option = [int(opt.strip()) for opt in input_option_str.split(',')]
            else:
                input_option = []
          else:
            print('해당 메뉴에 대한 선택 가능한 옵션이 없습니다.')
            input_option = [] # 선택 가능한 옵션이 없으면 input_option을 빈 리스트로 초기화

          # 선택한 옵션에 대한 유효성 검증
          if lenstr(input_option) > 0:
            set_input_option = set(input_option)
            unique_input_option = list(set_input_option)
            # 옵션을 오름차순으로 정렬, 주문 가능한 옵션만 저장
            for num in unique_input_option:
              if num in tmp_option:
                tmp_select.append(num)
                price_option += self.option[num][1]
              else:
                print(f"선택한 {num}옵션은 해당 메뉴에서 선택 불가하므로 주문에서 제외됩니다.")
        else:
          print('메뉴번호가 잘못되었습니다. 다시 입력해주세요')
      except ValueError:
        print('잘못된 입력입니다. 숫자를 입력해주세요.')
      else:
        # 해당음료의 주문 수량을 받음
        is_success = False
        while is_success != True:
          try:
            order_quantity = int(input("해당 메뉴를 주문 수량을 입력하세요 >>> "))
          except ValueError:
            print('잘못된 입력입니다. 숫자만 입력해주세요.')
          else:
            # 첫번째 메뉴라면 5자리 주문번호를 생성 첫칸에 입력
            if i == 0:
              self.orderlist.append(int(random()*10**5))
            self.orderlist.append([mn_num, tmp_select, order_quantity])
            self.orderprice.append((price_basic + price_option) * order_quantity)
            is_success = True
            i += 1
      finally:
        is_end = input('주문을 끝내려면 y를 입력해주세요. >>> ')
        if is_end.lower() == 'y': # 대소문자 구분 없이 'y' 입력 시 종료
            pass # 종료
        else:
            is_end = 'n' # 'y'가 아니면 계속 주문
    print('- '*23)
    print('주문이 완료되었습니다.')
    print('- '*23)


  ## 지불메소드
  def pay(self):
    self.ordertotal = sum(self.orderprice)
    print(f"주문금액은 총 {self.ordertotal:,}원입니다.")

    t = 0
    t = int(input('현금 결제는 1번, 카드 결제는 2번을 입력해주세요'))
    if t == 1:
        self.paymethod = "cash"
        print("직원을 호출하겠습니다")
    elif t == 2:
        self.paymethod = "card"
        print("IC칩 방향에 맞게 카드를 꽂아주세요.")
    else:
        print("1과 2 중 하나를 입력하세요.\n")

  ## 주문내역 저장
  def saveOrder(self):
    try:
      self.orderstate = {0:'ORDERED',1:'ACCEPTED',2:'PREPARING',3:'READY',4:'COMPLETED'}
      self.now = datetime.now()
      tmp_order = {
          'datetime': self.now,
          'ordernum':self.orderlist[0],
          'orderlist':self.orderlist[1:],
          'ordertotal':self.ordertotal,
          'paymethod':self.paymethod,
          'state':self.orderstate[0]
      }
    except:
      print("주문 데이터에 오류가 있습니다.")
    else:
      self.orderhistory.append(tmp_order)
      self.idx += 1
      print(f"주문번호 {self.orderlist[0]}번이 저장되었습니다.")

  ## 영수증 출력
  def printBill(self):
    # 주문 리스트를 주문번호와 주문내용으로 나눔
    order_num = self.orderlist[0]
    order_list = self.orderlist[1:]

    # Hot/Ice 표기는 제일 앞에 표기되므로 옵션 분리하기 위한 변수 선언
    ice_key = None
    for key, value in option_info.items():
      if value[0] == 'Ice':
        icd_key = key
    is_ice_num = option_info

    # 영주증 헤더 출력
    rpt = 20
    print('─' * rpt)
    print('　' * int((rpt - 5)//2) + '영　수　증')
    print('─' * rpt)
    print('[주문번호]'+str(order_num))
    print()
    print('상품명' + '　' * 9 + '단 가' + '　' * 3 + '수 량')
    print('- ' * rpt)

    # 메뉴 출력
    for order in order_list:
      mn_main_str = ''
      mn_option_str = ''

      # 메뉴번호를 카테고리키와 메뉴키로 분리
      mn_num = int(order[0]) # 메뉴번호
      ct_key = int(str(mn_num)[0]) # 카테고리키
      mn_key = int(str(mn_num)[1:]) - 1 # 메뉴키(0-based 인덱스로 변경)
      mn_option = order[1]
      mn_quantity = order[2] # 주문 수량

      # 메뉴키로 메뉴이름을 불러와서 메인 출력 문구에 추가
      mn_name = str(menu_list[ct_key][mn_key][0])
      mn_temp = 'HOT' # 기본으로 뜨거운 음료. 생각해보니 only ice 음료도 있는데.... 수정해야하는 부분이구나...!
      mn_temp = 'ICE'
      mn_main_str += '(' + mn_temp + ')' + mn_name
      mn_main_str += ' ' *  max(0, rpt - lenstr(mn_name) - lenstr(mn_temp))

      # 메뉴키로 가격정보 불러와서 메인 출력 문구에 추가
      mn_basic_price = menu_list[ct_key][mn_key][1]
      mn_main_str += ' ' *  max(0, int((10 - lenstr(f"{mn_basic_price:,}"))/2))
      mn_main_str += str(f"{mn_basic_price:,}")

      # 주문수량을 출력 문구에 추가
      mn_main_str += ' ' *  max(0, 9 - lenstr(str(mn_quantity)))
      mn_main_str += str(mn_quantity) + '개'

      # 옵션출력하기
      if lenstr(order[1]) > 0:
        for option in mn_option:
          if option not in [0,1]:
            option_nm = self.option[option][0]
            option_price = self.option[option][1]
            mn_option_str += ' └ ' + option_nm
            mn_option_str += ' ' *  max(0, rpt + 4 - lenstr(option_nm) - lenstr(f"{option_price:,}"))
            mn_option_str += '+' + f"{option_price:,}" + '\n'

      print(mn_main_str)
      print(mn_option_str)

    print('- ' * rpt)
    formated_date = self.now.strftime("%Y-%m-%d %H:%M:%S")
    mn_time_str = '주문시간'
    mn_time_str += ' ' *  max(0, rpt*2 - 9 - lenstr(formated_date))
    mn_time_str += formated_date
    mn_price_str = '총주문금액'
    mn_price_str += ' ' *  max(0, rpt*2 - 12 - lenstr(f"{self.ordertotal:,}"))
    mn_price_str += f"{self.ordertotal:,}원"
    mn_pay_str = '결제수단'
    mn_pay_str += ' ' *  max(0, rpt*2 - 9 - lenstr(self.paymethod))
    mn_pay_str += self.paymethod
    print(mn_time_str)
    print(mn_price_str)
    print(mn_pay_str)
    print()
    print('─' * rpt)
```
