## **1차 코드 작성**

### **[퀘스트 문제 확인]**
#### Q1. Account 클래스
- 은행이름, 예금주, 계좌번호, 잔액 설정
- 생성자에서는 예금주와 초기 잔액만 입력받기
- 은행이름은 SC은행
- 계봐번호는 3자리-2자리-6자리 형태로 랜덤하게 
- 출력은 **은행이름: SC은행, 계좌번호: 111-11-111111 **

### **[코드 작성 방향]**
- 은행이름은 클래스 변수
- 은행마다 계좌변호 형식이 다를 수 있기에 클래스 변수로 계좌형식 변수 생성
- 계좌는 중복되면 안되기에 클래스변수로 계좌set을 만들어서 생성된 계좌번호가 기존 set에 있는지 확인
  -> 계좌생성기 별도 작성
- 계좌번호를 생성하기 위해 random 임포트 필요

### **1차 작성 코드**

````python
from random import random # 랜덤 함수 임포트
````

````python
class Account_vr01:
    bank_nm = 'SC은행'     # 은행명
    bank_ac_list = set()   # 계좌번호 set. (중복되면 안되므로 set으로 설정)
    ac_format = [3,2,6]    # 계좌번호 포맷 지정  
    # 객체 생성기
    def __init__(self,ac_user,inti_bal):
        self.ac_user = ac_user        # 예금주 
        self.ac_bal = inti_bal      # 초기잔액을 잔액으로 저장
        self.ac_num = self.generate_ac()   # 계좌번호 할당
        self.bank_ac_list.add(self.ac_num) # 계좌번호를 Set에 추가

    # 계좌번호 생성기
    def generate_ac(self):
        ac_num_temp = ''     # 계좌번호를 임시 저장할 변수
        is_success = False
        while is_success != True:            
            for i in self.ac_format:
                if ac_num_temp != '':
                    ac_num_temp += '-'
                ac_num_temp += str(random())[2:2+i]
            if ac_num_temp not in self.bank_ac_list:
                is_success = True
            else:
                ac_num_temp = ''   # 생성된 계좌번호 지우기
        return ac_num_temp    
````

````python
class Account_vr03:
    bank_nm = 'SC은행'     # 은행명
    bank_ac_list = set()   # 계좌번호 set. (중복되면 안되므로 set으로 설정)
    ac_format = [3,2,6]    # 계좌번호 포맷 지정  
   
    
    # 계좌 생성기
    def __init__(self,ac_user,inti_bal):
        self.ac_user = ac_user        # 예금주 
        self.ac_bal = inti_bal      # 초기잔액을 잔액으로 저장
        self.ac_num = self.generate_ac()   # 계좌번호 할당
        self.bank_ac_list.add(self.ac_num) # 계좌번호를 Set에 추가
        self.ac_seq = 0
        print("신규계좌: "+ self.ac_num + '　'*8 + ' '*(29 - len(self.ac_num) - len(f"{self.ac_bal:,}")) + f"{self.ac_bal:,}" +"원")

    # 계좌번호 생성기
    def generate_ac(self):
        ac_num_temp = ''     # 계좌번호를 임시 저장할 변수
        is_success = False
        while is_success != True:            
            for i in self.ac_format:
                if ac_num_temp != '':
                    ac_num_temp += '-'
                ac_num_temp += str(random())[2:2+i]
            if ac_num_temp not in self.bank_ac_list:
                is_success = True
            else:
                ac_num_temp = ''   # 생성된 계좌번호 지우기
        return ac_num_temp    

    # 이번 항목에서 테스트하지 않을 메소드는 임시 삭제
    
    # 입금 메서드
    def deposit(self,amount,remark=""):
        try:
            self.ac_bal += amount
        except:
            print("입금 처리되지 않았습니다.")
        else:
            self.ac_seq += 1
            temp_space_1 = ' '*(5 - len(str(self.ac_seq)))
            if remark != "" and len(remark) > 8:
                temp_remark = remark[:8]
            else:
                temp_remark = remark
            
            temp_space_1 +=  '　'*(10 - len(temp_remark))
            temp_amount = f"{amount:,}"
            temp_space_2 = ' '*(10 - len(temp_amount))
            temp_balance = f"{self.ac_bal:,}" + "원"
            temp_space_3 = '   잔액:' +' '*(10 - len(temp_balance))
            print( str(self.ac_seq) + temp_space_1 + temp_remark + ' + ' +temp_space_2 + temp_amount + temp_space_3  + temp_balance)
    # 출금 메서드
    def withdraw(self,amount,remark=""):
        try:
            if amount > self.ac_bal:
                print("출금 가능 금액이 모자랍니다.")
            else:
                self.ac_bal -= amount 
        except:
            print("출금 처리되지 않았습니다.")
        else:
            self.ac_seq += 1
            temp_space_1 = ' '*(5 - len(str(self.ac_seq)))
            if remark != "" and len(remark) > 8:
                temp_remark = remark[:8]
            else:
                temp_remark = remark
            
            temp_space_1 +=  '　'*(10 - len(temp_remark))
            temp_amount = f"{amount:,}"
            temp_space_2 = ' '*(10 - len(temp_amount))
            temp_balance = f"{self.ac_bal:,}" + "원"
            temp_space_3 = '   잔액:' +' '*(10 - len(temp_balance))
            print( str(self.ac_seq) + temp_space_1 + temp_remark + ' - ' +temp_space_2 + temp_amount + temp_space_3  + temp_balance)
````

````python
from random import random, randint
````

````python
class Account_vr05:
    bank_nm = 'SC은행'     # 은행명
    bank_ac_list = set()   # 계좌번호 set. (중복되면 안되므로 set으로 설정)
    ac_format = [3,2,6]    # 계좌번호 포맷 지정  
   
    
    # 계좌 생성기
    def __init__(self,ac_user,inti_bal):
        self.ac_user = ac_user        # 예금주 
        self.ac_bal = inti_bal      # 초기잔액을 잔액으로 저장
        self.ac_num = self.generate_ac()   # 계좌번호 할당
        self.bank_ac_list.add(self.ac_num) # 계좌번호를 Set에 추가
        self.ac_seq = 0
        self.ac_seq_in = 0
        self.ac_seq_out = 0
        self.ac_deposit_history = []
        self.ac_withdraw_history = []
        self.ac_log = []
        tmp_msg = "[신규계좌]"+ self.ac_num + '　'*8 + ' '*(34 - len(self.ac_num) - len(f"{self.ac_bal:,}")) + f"{self.ac_bal:,}" +"원"
        print(tmp_msg)

    # 계좌번호 생성기
    def generate_ac(self):
        ac_num_temp = ''     # 계좌번호를 임시 저장할 변수
        is_success = False
        while is_success != True:            
            for i in self.ac_format:
                if ac_num_temp != '':
                    ac_num_temp += '-'
                ac_num_temp += str(random())[2:2+i]
            if ac_num_temp not in self.bank_ac_list:
                is_success = True
            else:
                ac_num_temp = ''   # 생성된 계좌번호 지우기
        return ac_num_temp    

    # 개설된 개좌 갯수 출력기
    def get_account_num(self):
        print("생성된 계좌의 수는 모두 " + str(len(self.bank_ac_list)) + "입니다")
        return True
            
    # 계좌 정보 출력기
    def display_info(self):
        str_temp = '('
        str_temp += '은행이름: ' + self.bank_nm
        str_temp += ', 예금주: ' + self.ac_user
        str_temp += ', 계좌번호: ' + self.ac_num
        str_temp += ', 잔고: ' + ' '*(12 - len(f"{self.ac_bal:,}")) + f"{self.ac_bal:,}" + '원'
        str_temp += ')' 
        print(str_temp)
        return True
    
    # 입금 메서드
    def deposit(self,amount,remark=""):
        try:
            self.ac_bal += amount
        except:
            print("입금 처리되지 않았습니다.")
        else:
            self.ac_seq += 1
            self.ac_seq_in += 1
            temp_space_1 = ' '*(5 - len(str(self.ac_seq)))
            if remark != "" and len(remark) > 8:
                temp_remark = remark[:8]
            else:
                temp_remark = remark
            
            temp_space_1 +=  '　'*(10 - len(temp_remark))
            temp_amount = f"{amount:,}"
            temp_space_2 = ' '*(10 - len(temp_amount))
            temp_balance = f"{self.ac_bal:,}" + "원"
            temp_space_3 = '   잔액:' +' '*(15 - len(temp_balance))            
            tmp_msg = str(self.ac_seq) + temp_space_1 + temp_remark + ' + ' +temp_space_2 + temp_amount + temp_space_3  + temp_balance
            print(tmp_msg)
            self.ac_deposit_history.append([self.ac_seq,remark,amount,self.ac_bal])

            if self.ac_seq_in % 5 == 0 :
                self.apply_interest()

    # 이자 지급기
    def apply_interest(self):
        temp_interest = int(self.ac_bal * 0.01)
        try:
            self.ac_bal += temp_interest
        except:
            print("이자지급처리가 제대로 되지 않았습니다.")
            print("은행 관리자에게 문의하세요")
            is_success = False
        else:
            self.ac_seq += 1
            temp_space_1 = ' '*(5 - len(str(self.ac_seq)))
            temp_remark = "이자지급"
            temp_space_1 +=  '　'*(10 - len(temp_remark))
            temp_amount = f"{temp_interest:,}"
            temp_space_2 = ' '*(10 - len(temp_amount))
            temp_balance = f"{self.ac_bal:,}" + "원"
            temp_space_3 = '   잔액:' +' '*(15 - len(temp_balance))
            tmp_msg = str(self.ac_seq) + temp_space_1 + temp_remark + ' + ' +temp_space_2 + temp_amount + temp_space_3  + temp_balance
            print(tmp_msg)
            self.ac_deposit_history.append([self.ac_seq,remark,amount,self.ac_bal])

            is_success = True
        return is_success                    
            
    # 출금 메서드
    def withdraw(self,amount,remark=""):
        try:
            if amount > self.ac_bal:
                print("출금 가능 금액이 모자랍니다.")
            else:
                self.ac_bal -= amount 
        except:
            print("출금 처리되지 않았습니다.")
        else:
            self.ac_seq += 1            
            self.ac_seq_out += 1
            temp_space_1 = ' '*(5 - len(str(self.ac_seq)))
            if remark != "" and len(remark) > 8:
                temp_remark = remark[:8]
            else:
                temp_remark = remark
            
            temp_space_1 +=  '　'*(10 - len(temp_remark))
            temp_amount = f"{amount:,}"
            temp_space_2 = ' '*(10 - len(temp_amount))
            temp_balance = f"{self.ac_bal:,}" + "원"
            temp_space_3 = '   잔액:' +' '*(15 - len(temp_balance))   
            tmp_msg = str(self.ac_seq) + temp_space_1 + temp_remark + ' - ' +temp_space_2 + temp_amount + temp_space_3  + temp_balance
            print(tmp_msg)
            self.ac_withdraw_history.append([self.ac_seq,remark,amount,self.ac_bal])
            
    # 입금 기록
    def deposit_history(self):
        self.display_info()
        print('[입금기록]')
        for record in self.ac_deposit_history:
            rcd_seq,rcd_remark,rcd_amount,rcd_balance = record
            if len(rcd_remark) > 8:
                tmp_remark = rcd_remark[:8]
            else:
                tmp_remark = rcd_remark
            tmp_space_1 = ' '*(5 - len(str(rcd_seq))) + '　'*(10 - len(tmp_remark))
            tmp_amount = f"{rcd_amount:,}"
            tmp_space_2 = ' '*(10 - len(tmp_amount))
            tmp_balance = f"{rcd_balance:,}" + "원"
            tmp_space_3 = '   잔액:' +' '*(15 - len(tmp_balance))  
            tmp_msg = str(rcd_seq) + tmp_space_1 + tmp_remark + ' + ' +tmp_space_2 + tmp_amount + tmp_space_3  + tmp_balance
            print(tmp_msg)
        print("--"*30)
    # 출금 기록
    def withdraw_history(self):
        self.display_info()
        print('[출금기록]')
        for record in self.ac_withdraw_history:
            rcd_seq,rcd_remark,rcd_amount,rcd_balance = record
            if len(rcd_remark) > 8:
                tmp_remark = rcd_remark[:8]
            else:
                tmp_remark = rcd_remark
            tmp_space_1 = ' '*(5 - len(str(rcd_seq))) + '　'*(10 - len(tmp_remark))
            tmp_amount = f"{rcd_amount:,}"
            tmp_space_2 = ' '*(10 - len(tmp_amount))
            tmp_balance = f"{rcd_balance:,}" + "원"
            tmp_space_3 = '   잔액:' +' '*(15 - len(tmp_balance))  
            tmp_msg = str(rcd_seq) + tmp_space_1 + tmp_remark + ' - ' +tmp_space_2 + tmp_amount + tmp_space_3  + tmp_balance
            print(tmp_msg)
        print("--"*30)
        
````
