## **1단계: Character 클래스 만들기**
가장 먼저 게임 내 캐릭터(플레이어, 몬스터)의 공통 속성을 담은 부모 클래스를 정의합니다.

```python
import random

class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        """현재 체력이 0 이상인지 확인합니다."""
        return self.hp > 0

    def take_damage(self, damage):
        """공격을 받았을 때 방어력을 고려해 체력을 감소시킵니다."""
        actual_damage = max(0, damage - self.defense)
        self.hp -= actual_damage
        print(f"{self.name}이(가) {actual_damage}의 데미지를 입었습니다. (남은 체력: {max(0, self.hp)})")

    def attack_target(self, target):
        """1부터 본인의 공격력 사이의 랜덤한 데미지를 입힙니다."""
        damage = random.randint(1, self.attack)
        print(f"{self.name}이(가) {target.name}을(를) 공격합니다! (공격력: {damage})")
        target.take_damage(damage)
```

## **2단계: Player 및 Monster 클래스 만들기**
Character를 상속받아 각각의 특화된 속성과 메서드를 추가합니다.

```python
class Player(Character):
    def __init__(self, name):
        # 이름은 입력받고 나머지는 기본값으로 초기화
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.exp = 0

    def gain_experience(self, amount):
        """경험치를 획득하고 레벨업 여부를 확인합니다."""
        self.exp += amount
        print(f"{amount}의 경험치를 획득했습니다. (현재 경험치: {self.exp})")
        self.level_up()

    def level_up(self):
        """경험치가 50 이상이면 레벨과 능력치가 상승합니다."""
        if self.exp >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.hp = 100 + (self.level * 10)  # 레벨업 시 체력 회복 및 최대치 증가 (선택 사항)
            self.exp -= 50
            print(f"★ LEVEL UP! ★ 레벨 {self.level}이 되었습니다!")
            print(f"공격력: {self.attack}, 방어력: {self.defense}")

class Monster(Character):
    def __init__(self, name, level):
        # 레벨에 비례하여 랜덤한 능력치 생성
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)
        print(f"몬스터 '{self.name}'(Lv.{self.level})가 나타났습니다!")
```

## **3단계: 전투(Battle) 함수 만들기**
플레이어와 몬스터가 서로 죽을 때까지 공방을 주고받는 핵심 로직입니다.

```python
def battle(player, monster):
    print(f"\n--- {monster.name}와(과)의 전투 시작! ---")
    
    while player.is_alive() and monster.is_alive():
        # 플레이어의 턴
        player.attack_target(monster)
        if not monster.is_alive():
            print(f"\n{monster.name}을(를) 처치했습니다!")
            player.gain_experience(monster.level * 20)
            print("전투 승리!")
            break
            
        # 몬스터의 턴
        monster.attack_target(player)
        if not player.is_alive():
            print("\n플레이어가 쓰러졌습니다...")
            print("전투 패배..")
            break
```

## **4단계: 게임 실행 (Main)**
사용자로부터 이름을 입력받고 실제로 게임을 구동하는 부분입니다.

```python
# 1. 플레이어 생성
user_name = input("플레이어의 이름을 입력하세요: ")
player = Player(user_name)

# 2. 스테이지 진행 (예: 3마리의 몬스터와 차례로 전투)
monster_names = ["슬라임", "고블린", "오크", "드래곤"]

for i in range(len(monster_names)):
    if not player.is_alive():
        break
    
    # 몬스터 레벨은 진행도에 따라 증가하도록 설정
    monster = Monster(monster_names[i], level=i+1)
    battle(player, monster)
    print("-" * 30)

print("\n게임이 종료되었습니다.")
```
